```python
import pandas as pd
df  = pd.read_csv("/content/data.csv")
df.head(50)
```

```python
df.columns
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn import metrics
from flask import Flask, render_template, jsonify



X = df.drop(columns = ['filename','label'], axis=1)

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X_scaled)

# Get cluster labels
cluster_labels = kmeans.labels_

# Evaluate the clustering using Silhouette Coefficient
silhouette_score = metrics.silhouette_score(X_scaled, cluster_labels)
print("Silhouette Coefficient:", silhouette_score)
```

```python

# Flask App
app = Flask(__name__)

@app.route('/')
def index():
    # Display clustering results in frontend
    return render_template('/content/index.html')

@app.route('/get_cluster_labels')
def get_cluster_labels():
    # Return cluster labels as JSON
    return jsonify({'cluster_labels': cluster_labels.tolist()})

@app.route('/get_silhouette_score')
def get_silhouette_score():
    # Return silhouette score as JSON
    return jsonify({'silhouette_score': silhouette_score})

if __name__ == '__main__':
    app.run()
```
